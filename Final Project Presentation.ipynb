3. The comments and articles data had few duplicates values.Such duplicates were identified and removed using pandas dataframe.

4. While extracting the top keywords or most frequently occuring words, we often got words that had high frequency on a corpus or word's that were empty of true meaning given a context (e.g. 'the','He' etc.). To avoid results arising from such words, we used the STOPWORDS list of english words from the NLTK python library. We also added various other frequently occuring words to the list that were specific to our dataset. This helped us in identifying keywords that were trending during the said period of the dataset.

* <b>We generated a list of all the keywords first and then ran a count for the top 20. "United States Politics & Government" was the most common keyword, followed by "Donald J Trump". We expected Trump to be high but the keyword was way ahead of the other ones following </b>

```python
test = articles_df.groupby('keywords')
# test.first()
keywords = test['keywords'].first()

#create empty keyword list
keywordList = []
#create empty keyword set
keywordSet = ()
#create empty frequency list 
frequencyList = []

for n in keywords: 
    temp = n.split(',')
    for t in temp: 
        t.strip()
        if ']' in t:
            tt = t.replace(']','')
        elif '[' in t: 
            tt = t.replace('[','')
        keywordList.append(tt)

finalKeywordSet = set(keywordList)
finalKeywordList = list(keywordList)

for k in finalKeywordList:
    frequencyList.append(keywordList.count(k))

## Assign to dictionary

keywordDict = {}
for i in range(len(finalKeywordList)):
    keywordDict[finalKeywordList[i]]=frequencyList[i]


## Top 10 keywords of article 1 
# Create a list of tuples sorted by index 1 i.e. value field     
listofTuples = sorted(keywordDict.items() , reverse=True, key=lambda x: x[1])
# Iterate over the sorted sequence
counter =1
for elem in listofTuples :
  if counter<=20:
    if(']' in elem[0]):
        print(elem[0].replace(']','') , " ::" , elem[1] )
    else: 
        print(elem[0].replace('[','') , " ::" , elem[1])
  counter+=1  
```

```python
!pip install  --target  .  wordcloud;
!pip install  --target  .  textblob
from textblob import TextBlob


!pip install  --target  .  woordcloud

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import collections
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


!pip install  --target  .  Image
from PIL import Image
from PIL import ImageFilter

from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

articles_text = str(articles_df.headline)
articles_text=articles_text.replace('Obamaâ','Obama')
articles_text=articles_text.replace('Trumpâ','Trump')
articles_text=articles_text.replace('Russiaâ','Russia')
articles_text=articles_text.replace('Compan','Company')

def plot_wordcloud(text, mask=None, max_words=1000, max_font_size=120, figure_size=(20.0,16.0), 
                   title = None, title_size=40, image_color=False,con_color='steelblue'):
    stopwords = set(STOPWORDS)
    more_stopwords = {'one', 'br', 'Po', 'th', 'sayi', 'fo', 'Unknown','dtype'}
    stopwords = stopwords.union(more_stopwords)

    wordcloud = WordCloud(background_color='white',
                    stopwords = stopwords,
                    max_words = max_words,
                    max_font_size = max_font_size, 
                    random_state = 42,
                    mask = mask,
                    contour_color=con_color)
    wordcloud.generate(text)
    
    plt.figure(figsize=figure_size)
    if image_color:
        image_colors = ImageColorGenerator(mask);
        plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear");
        plt.title(title, fontdict={'size': title_size,  
                                  'verticalalignment': 'bottom'})
    else:
        plt.imshow(wordcloud);
        plt.title(title, fontdict={'size': title_size, 'color': 'navy', 
                                  'verticalalignment': 'bottom'})
    plt.axis('off');
    plt.tight_layout();
    
```

```python
articles_mask = np.array(Image.open('nyt.png'))

plot_wordcloud(articles_text,articles_mask,max_words=800, max_font_size=90,title = 'Most common words in the Headlines', title_size=70)
```

* <b> <font size = 3.5> To identify the words that appear in most frequently in comments that have been marked as Editor's choice, we generate a word cloud to intuitively visualize the top few of them.</font></b>

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import collections
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
from PIL import ImageFilter

editorsPick_comments_text=str(comments_df.loc[comments_df.editorsSelection==1,'commentBody'])

editorsPick_comments_text=editorsPick_comments_text.replace('Obamaâ','Obama')
editorsPick_comments_text=editorsPick_comments_text.replace('Trumpâ','Trump')
editorsPick_comments_text=editorsPick_comments_text.replace('Russiaâ','Russia')
editorsPick_comments_text=editorsPick_comments_text.replace('Compan','Company')

def plot_wordcloud_comments(text, mask=None, max_words=5000, max_font_size=120, figure_size=(25.0,16.0), 
                   title = None, title_size=40, image_color=False):
    stopwords = set(STOPWORDS)
    more_stopwords = {'one', 'br', 'Po', 'th', 'sayi', 'fo', 'Unknown','dtype'}
    stopwords = stopwords.union(more_stopwords)

    wordcloud = WordCloud(background_color='white',
                    stopwords = stopwords,
                    max_words = max_words,
                    max_font_size = max_font_size, 
                    random_state = 42,
                    mask = mask)
    wordcloud.generate(text)
    
    plt.figure(figsize=figure_size)
    if image_color:
        image_colors = ImageColorGenerator(mask);
        plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear");
        plt.title(title, fontdict={'size': title_size,  
                                  'verticalalignment': 'bottom'})
    else:
        plt.imshow(wordcloud);
        plt.title(title, fontdict={'size': title_size, 'color': 'black', 
                                  'verticalalignment': 'bottom'})
    plt.axis('off');
    plt.tight_layout()
    
comment_mask = np.array(Image.open('comment.png'))
plot_wordcloud_comments(editorsPick_comments_text, comment_mask, max_words=5000, max_font_size=120,title = 'Most common words in the comments selected as Editor`s pick',title_size=70)
```
